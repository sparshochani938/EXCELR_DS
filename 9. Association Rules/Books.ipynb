```python
from mlxtend.frequent_patterns import apriori,association_rules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('book.csv')
data
```

# Associating when support is 1% and confidence is 70%

```python
frequent_data=apriori(data,min_support=0.1,use_colnames=True)
frequent_data
```

```python
rules=association_rules(frequent_data,metric='lift',min_threshold=0.7)
```

```python
rules[rules.lift>1]
plt.scatter(rules['support'],rules['confidence'])
plt.title('S=1% and C=70%')
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

# B. Assosiating when S is 20% and C is 90%

```python
frequent_data=apriori(data,min_support=0.2,use_colnames=True)
rules=association_rules(frequent_data,metric='lift',min_threshold=0.9)
plt.scatter(rules['support'],rules['confidence'])
plt.title('S=20% and C=90%')
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
